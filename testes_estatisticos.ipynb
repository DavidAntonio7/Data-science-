```python
#intervalo de confianca para o mundo real e nao para o conjunto de dados
zconfint(notas1.rating)
```

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating,value=3.432) #valor da media das notas

#como a respota 8.909392288271546e-18 e' menor que 5% nao e' correto
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(75421)
temp = notas1.sample(frac =1).rating

medias = [temp[0:i].mean()for i in range(1,len(temp))]

plt.plot(medias)
plt.title('notas aleatorias de medias do toy story')
```

No grafico abaixo para 25 notas mostra que o media do toy story pode ser considera 3.43, um erro que pequenas amostras podem causar

```python
np.random.seed(75421)
temp = notas1.sample(frac =1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stat,p = ztest(temp[0:i],value = 3.432)
  return (i,media,p)

valores = np.array([calcula_teste(i) for i in range(2,len(temp))])

plt.plot(valores[:,0],valores[:,1])
plt.plot(valores[:,0],valores[:,2])
plt.hlines(y=0.05,xmin=2,xmax = len(temp),colors = 'g')
plt.title('notas aleatorias de medias e o pvalue do toy story')
```
